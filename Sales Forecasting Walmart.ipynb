```python
#Step-1 First, importing libraries of Python.
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from scipy.special import boxcox1p
import seaborn as sns
import warnings
```

```python
#Step-2 Now, we preparing data
#Getting data from .csv file
#Getting data
features=pd.read_csv("features.csv")
store=pd.read_csv("stores.csv")
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
```

```python
'''Take important information and merging the data 
   From train.csv, taking groupby Store, Date and sum Weekly_Sales.
'''
#reseting train
train=train.groupby(['Store','Date'])['Weekly_Sales'].sum()
train=train.reset_index()
train.head(10)
```

```python
#Merging train and features data by inner join.

data=pd.merge(train,features,on=['Store','Date'],how='inner')
data.head(10)
```

```python
#Merging Data and Store on Store by inner join.
#marging store with data
data=pd.merge(data,store,on=['Store'],how='inner')
data.head(10)
```

```python
#sorting the data by date.
data=data.sort_values(by='Date')
data.head(10)
```

```python
#Here, we see different methods to analyze data.
#Count plot of Type.
sns.countplot(x="Type", data=data)
```

```python
#Histogram of Weekly_sales
data["Weekly_Sales"].plot.hist()
```

```python
#Count plot of IsHoliday
sns.countplot(x="IsHoliday", data=data)
```

```python
#next check the null values of data
data.isnull().sum()
```

```python
#Removing the unnecessary columns.
data=data.drop(['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5'],axis=1)
data.head(10)
```

```python
#checking null values in data
data.isnull().sum()
```

```python
#Converting IsHoliday in Holiday which is integer and 1 for holiday and 0 otherwise.
data['Holiday']=[int(i) for i in list(data.IsHoliday)]
data.head(10)
```

```python
'''The first column is also removed because we know of both columns B and C are 0 then it is A-type.
   So B=1 and C=0 for B.B=0 and C=1 for C.B=0 and C=0 for A.
'''
Type_dummy=pd.get_dummies(data['Type'],drop_first=True)
Type_dummy.head(10)
```

```python
#Concating type_dummy with data.
data=pd.concat([data,Type_dummy],axis=1)
data.head(10)
```

```python
#STEP-3
#Splitting data into train and test data. The size of the test data is 30%.
#splitting data in input and output
X=data.drop(['Weekly_Sales','Store','Date'],axis=1)
y=data['Weekly_Sales']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
#Applying linear regression and fit the training data into it.
LR=LinearRegression()
LR.fit(X_train,y_train)
```

```python
#Predicting the data for test value as per linear regression.

y_pred=LR.predict(X_test)
plt.plot(y_test,y_pred,'ro')
plt.plot(y_test,y_test,'b-')
plt.show()
```

```python
#Evaluating the model by calculating errors by the root mean square error and R -squared.

Root_mean_square_error=np.sqrt(np.mean(np.square(y_test-y_pred)))
print(Root_mean_square_error)
```

```python
#calculating accuracy
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print(r2)
```
